# PRE TRAINED MODELS

```python
from keras.applications import VGG16
from  tensorflow.keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
```

```python
# Load the VGG16 pre-trained model
model = VGG16(weights='imagenet', include_top=True, input_shape=(224, 224, 3))
```

```python
# Loading an image and preprocess it for model
import numpy as np
img_path = '2295034.webp'
img = image.load_img(img_path,target_size=(224,224))
X = image.img_to_array(img)
X = np.expand_dims(X, axis=0)
X = preprocess_input(X)
```

```python
# Use the model to predict the class
preds = model.predict(X)
```

```python
preds
```

```python
#Get the top 5 prediction with class names
decode_preds = decode_predictions(preds,top=10)[0]
```

```python
for pred in decode_preds:
    print(pred[1], ":", pred[2])
```
